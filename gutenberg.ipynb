```python
authors = pd.read_csv(
    "https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-06-03/gutenberg_authors.csv"
)

authors.head()

#authors.columns
```

```python
meta = pd.read_csv(
    "https://raw.githubusercontent.com/rfordatascience/tidytuesday/main/data/2025/2025-06-03/gutenberg_metadata.csv")

meta.head()

#meta.columns
```

```python
df = pd.merge(
    authors,
    meta,
    on="gutenberg_author_id",
    how="inner"
)

df.head()
```

```python
df["language"].unique()
df["language"].value_counts()
```

```python
alias_counts = (
    df
    .groupby("alias")
    .size()
    .sort_values(ascending=False)
)
```

```python
list_authors = alias_counts.index.tolist()
list_authors[:10]
```
